**The difference between re.search and re.match is that the matched substring of re.search does not have to start from the beginning of the input string**

```python
regex = "science\s\w+\s"
searched = re.search(regex, "data science 101 was fun")
print(searched)
```

```python
searched.span()
```

```python
searched.group()
```

**re.finditer**

```python
#match 'data' followed by whitespace char, followed by atleast one alphanumeric char followed by whitespace char
regex = "data\s[\w]+\s"
for matched in re.finditer(regex, "data science 101 was fun! I love data mining and exploration"):
    print(matched.span())
    print(matched.group())
```

**The only difference between re.findall and re.finditer is that re.findall returns a list instead of an iterator and contains matched substrings instead of re.Match objects.**

```python
regex = "data\s\w+\s"
for matched in re.findall(regex, "data science 101 was fun! I love data mining and exploration"):
    print(matched)
```

```python
matched_list1 = re.findall(regex, "data science 101 was fun! I love data mining and exploration")
matched_list1
```

```python
#substituting data science with machine learning
regex = "data\s\w+\s"
change = re.sub(regex, 'machine learning ', "data science 101 was fun! I love data mining and exploration")
change
```

```python
#match the words after 'data'
regex = "data\s(\w+)\s"
for matched in re.findall(regex, "data science 101 was fun! I love data mining and exploration"):
    print(matched)
```
