**The goal is to successfully webscrape the page for the melee top 100 players and compare the number of mains across different characters**

```python
#importing packages
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.ssbwiki.com/Top_100_Melee_Players_of_All_Time'

data = requests.get(url)

soup = BeautifulSoup(data.text, 'html.parser')
```

```python
#getting player from table marked 'wikitable'
player = []
table = soup.find('table', {'class' : 'wikitable'})
for tr in table.find_all('tr'):
    values = [td.text.strip() for td in tr.find_all('td')]
    player.append(values)
```

```python
#so the list player gives me the rank and the player, but I still need the characters. It looks like the characters can be found by looking at the <a> tag 
# and choosing the title
for tr in table.find_all('tr'):
    for td in tr.find_all('td'):
        test2 = [tag for tag in td.find_all('a')]
```
