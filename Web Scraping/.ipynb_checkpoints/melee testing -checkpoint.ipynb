```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.ssbwiki.com/Top_100_Melee_Players_of_All_Time'

data = requests.get(url)

soup = BeautifulSoup(data.text, 'html.parser')
```

```python
table = soup.find('table',{'class':'wikitable'})
tbody = table.find('tbody')
```

```python
df = pd.DataFrame(columns = ['rank','player','empty1','empty2','empty3'])
```

```python
for row in tbody.find_all('tr'):
    columns = row.find_all('td')
    
    if (columns != []):
        rank = columns[0].text.strip()
        player = columns[1].text.strip()
        empty1 = columns[2].text.strip()
        empty2 = columns[3].text.strip()
        empty3 = columns[4].text.strip()
        
        df = df.append({'rank': rank, 'player': player, 'empty1': empty1, 'empty2': empty2, 'empty3': empty3}, ignore_index = True)
        
```

```python
df2 = pd.DataFrame(columns = ['rank','player'])

for row in tbody.find_all('tr'):
    columns = row.find_all('td')
    
    if (columns != []):
        rank = columns[0].text.strip()
        player = columns[1].text.strip()
        
        df2 = df2.append({'rank': rank, 'player': player}, ignore_index = True)
```

```python
df3 = pd.DataFrame(columns = ['main', 'href'])
for tr in tbody.find_all('tr'):
    for td in tbody.find_all('td'):
        character = td.find_all('a')
        
        if (character != []):
            main = character[0]['title']
            href = character[0]['href']
            
            df3 = df3.append({'main':main, 'href': href}, ignore_index = True)
```

```python
df4 = pd.DataFrame(columns=['secondary'])

for tr in tbody.find_all('tr'):
    for td in tbody.find_all('td'):
        character = td.find_all('a')
        
        if (character != []):
            secondary = character
            
            df4 = df4.append({'secondary' : secondary}, ignore_index = True)
```

```python
#what is the output of using just character in the if statement?
df4
```

So for having character in the if statement, for each td, I get a list that will contain text. so in this case, each player has a flag and then a separate <a> for their name all in one td. So I get a single entry into the data frame being empty (because the <a> for flag has no text) and their name (since the <a> for name does have text). Then I get the next td entry, which has information on the characters the player plays as. Since Mango has 4 characters, the next element in the dataframe has 4 empty slots (because each <a> for each character does not contain text)

```python
df4 = pd.DataFrame(columns=['secondary'])

for tr in tbody.find_all('tr'):
    for td in tbody.find_all('td'):
        character = td.find_all('a')
        
        if (character != []):
            secondary = character
            
            df4 = df4.append({'secondary' : secondary}, ignore_index = True)
```
