```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.ssbwiki.com/Top_100_Melee_Players_of_All_Time'

data = requests.get(url)

soup = BeautifulSoup(data.text, 'html.parser')
```

```python
new_soup = soup.find(class_ = 'wikitable')
```

```python
tbody = new_soup.find('tbody')
```

```python
df = pd.DataFrame(columns = ['rank','player','empty1','empty2','empty3'])
```

```python
for row in tbody.find_all('tr'):
    columns = row.find_all('td')
    
    if (columns != []):
        rank = columns[0].text.strip()
        player = columns[1].text.strip()
        empty1 = columns[2].text.strip()
        empty2 = columns[3].text.strip()
        empty3 = columns[4].text.strip()
        
        df = df.append({'rank': rank, 'player': player, 'empty1': empty1, 'empty2': empty2, 'empty3': empty3}, ignore_index = True)
```
