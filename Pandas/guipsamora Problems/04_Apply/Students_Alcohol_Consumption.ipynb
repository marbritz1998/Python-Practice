### **Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old**

```python
#so I am making a function that returns True or False to a column called legal_drinker if the age is over 17

#lets try it my way first
for index in df.index:
    if df.loc[index,'age'] > 17:
        df.loc[index,'legal_drinker'] = True
    else:
        df.loc[index,'legal_drinker'] = False
df.head()       
```

```python
#so lets make make a function
def majority(age_list):
    legal_drinker = []
    for age in age_list:
        if age > 17:
            legal_drinker.append(True)
        else:
            legal_drinker.append(False)
    return legal_drinker       
```

```python
df['legal_drinker'] = majority(df.age)
```

```python
df.head()
```

```python
#both ways work
```


<div class="alert alert-block alert-warning">
   Another way to do this

```python
def majority(x):
    if x > 17:
        return True
    else:
        return False
```

```python
df['legal_drinker'] = df['age'].apply(majority)
df.head()
```

```python
def times10(x):
    if type(x) is int:
        return 10 * x
    return x
```
