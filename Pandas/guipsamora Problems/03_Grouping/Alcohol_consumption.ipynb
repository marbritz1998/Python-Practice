# **Ex - GroupBy** 

### **Step 1. Import the necessary libraries**

```python
import pandas as pd
import numpy as np
```

### **Step 2. Import the dataset**

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv'
```

### **Step 3. Assign it to a variable called drinks**

```python
drinks = pd.read_csv(url)
```

```python
drinks.head()
```

### **Step 4. Which continent drinks more beer on average?**

```python
#badly worded question, I will do a groupby based on continent with the average liters of pur alcohol
drinks.groupby('continent').agg({'beer_servings':'mean'}).sort_values('beer_servings', ascending = False)
```

```python
#alternative way with pivot_table
pd.pivot_table(drinks, index = 'continent', values = 'beer_servings', aggfunc = 'mean').sort_values('beer_servings', ascending = False)
```

### **Step 5. For each continent print the statistics for wine consumption**

```python
drinks.groupby('continent')['wine_servings'].describe()
```

### **Step 6. Print the mean alcohol consumption per continent for every column**

```python
drinks.groupby(['continent']).agg({'total_litres_of_pure_alcohol':'mean'})
```

```python
drinks.groupby(['continent']).mean()
```

<div class="alert alert-block alert-warning">
    Needed to do the mean for all columns

### **Step 7. Print the median alcohol consumption per continent for every column**


```python
drinks.groupby(['continent']).agg({'total_litres_of_pure_alcohol':'median'})
```

```python
drinks.groupby('continent').median()
```

### **Step 8. Print the mean, min and max values for spirit consumption**

```python
drinks.groupby('continent').agg({'spirit_servings':['mean','min','max']})
```
