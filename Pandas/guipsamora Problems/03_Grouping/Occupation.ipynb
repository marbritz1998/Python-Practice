### **Step 1. Import the necessary libraries**

```python
import pandas as pd
import numpy as np
```

### **Step 3. Assign it to a variable called users**

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep = '|')
users.head()
```

### **Step 4. Discover what is the mean age per occupation**

```python
users.groupby('occupation').agg({'age':'mean'}).sort_values('age', ascending = False)
```

### **Step 5. Discover the Male ratio per occupation and sort it from the most to the least**

```python
# so the male ratio per occupation is    take the total users in the occupation and divide by the number of users who are male
occupation_grouped = users.groupby(['occupation'])
occupation_grouped.agg({'gender': 'count'})  #this gives the count of genders that are both male and female per occupation
```

```python
#so I am having troubles doing this by groupby, I could make another dataframe that is the subset with gender = male
```

```python
users_m = users[users['gender'] == 'M']
users_m.head()
```

```python
users_m.groupby(['occupation']).agg({'gender':'count'})  # this gives the count of genders that are male by occupation
```

```python
#this gives the male ratio per occupation
users_m.groupby(['occupation']).agg({'gender':'count'})/occupation_grouped.agg({'gender': 'count'})
```

```python
#this gives the male ratio per occupation in 1 line, lets sort fromt most to least
male_ratio = users[users['gender'] == 'M'].groupby(['occupation']).agg({'gender':'count'}) / users.groupby('occupation').agg({'gender':'count'})
```

```python
male_ratio.rename(columns={'gender':'Male-to-Female Ratio'}).sort_values('Male-to-Female Ratio', ascending = False)
```

### **Step 6. For each occupation, calculate the minimum and maximum ages**

```python
users.groupby('occupation').agg({'age':['max','min']})
```

### **Step 7. For each combination of occupation and gender, calculate the mean age**

```python
users.groupby(['occupation','gender']).agg({'age':'mean'})
```

### **Step 8. For each occupation present the percentage of women and men**

```python
male_ratio.rename(columns = {'gender':'Male-ratio'}, inplace = True)
male_ratio['Female-ratio'] = 1 - male_ratio['Male-ratio']
```

```python
occupation_ratio = male_ratio
occupation_ratio.style.format({'Male-ratio': "{:.2%}", 'Female-ratio': "{:.2%}"})  #this one was tough, I had to google for the formatter
```
